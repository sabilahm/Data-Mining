```python
#Tugas 1 - Data Integration

import pandas as pd
```

```python
ccfraud1 = pd.read_csv('creditcardfraud1.csv')
ccfraud2 = pd.read_csv('creditcardfraud2.csv')

ccf_merged = pd.merge(ccfraud1, ccfraud2, on = 'Time')
ccf_merged = pd.concat([ccfraud1, ccfraud2], sort=True)

ccf_merged.head()
```

```python
#Exporting
ccf_merged.to_csv("ccf_merged.csv")
```
