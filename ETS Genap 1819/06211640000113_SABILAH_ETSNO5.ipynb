```python
import pandas as pd
import numpy as np
from apyori import apriori
data = pd.read_csv("online_retail.csv")
data.head()
```

```python
np.sum(data.isnull())
```

```python
#memilih data hanya menggunakan data StockCode saja berdasarkan transaksi yang ada dilihat dari InvoiceNo
#dikarenakan terdapat missing value pada variabel Description dan CustomerID
#sehingga menggunakan InvoiceNo dan StockCode akan lebih bagus karena bisa menggambarkan seluruh transaksi yang ada
x = data.groupby('InvoiceNo')
data_s = []
for i,j in x:
    data_s.append(list(j['StockCode'].map(str)))
```

```python
pd.DataFrame(data_s)
```

```python
association_rules = apriori(data_s, min_support = 0.0045,
                           min_confidence = 0.2,
                           min_lift = 3,
                           min_length = 2)

association_result = list(association_rules)
```

```python
len(association_result)
```

```python
association_result[0]
```

```python
for item in association_result:
    pair = item[0]
    items = [x for x in pair]
    print("rule: " + items [0] + " ->" + items[1])
    print("support: " + str(items[1]))
    print("confidence: " + str(item[2][0][2]))
    print("lift: " + str(item[2][0][3]))
    print("==========")
```
